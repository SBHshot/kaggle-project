```python
import os
import random
import pandas as pd
from pydicom import dcmread
import matplotlib.pyplot as plt
```

```python
filePath = '/home/jun/python/dataset'
trainPath = '/home/jun/python/dataset/train'
testPath = '/home/jun/python/dataset/test'
```

```python
study_df = pd.read_csv(filePath+'/train_study_level.csv')
image_df = pd.read_csv(filePath+'/train_image_level.csv')
```

```python
image_df.head
```

```python
trainImgPath = []
uIdList = []
for idx,row in image_df.iterrows():
    uIdList.append(row['StudyInstanceUID'])
    dir = os.listdir(trainPath + "/" + row['StudyInstanceUID'])
    trainImgPath.append(trainPath+'/'+row['StudyInstanceUID']+'/'+dir[0]+'/'+row['id'][0:12]+'.dcm')
```

```python
x = random.choice(trainImgPath)
#print(x)
a = dcmread(x).pixel_array
plt.imshow(a,cmap='bone')
```
